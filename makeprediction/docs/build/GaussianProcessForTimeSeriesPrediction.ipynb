```python
%matplotlib inline
from gpasdnn.kernels import *
from gpasdnn.gp import GaussianProcessRegressor as gpr
from gpasdnn.quasigp import QuasiGPR as qgpr

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
T = 1000
x = np.linspace(0,10,T)
kernel = Periodic(period=2) + Periodic(period=.2) + Linear(variance=1) + RBF(1)
y = kernel.simulate(x,seed=np.random.seed(10))

iid = .2*np.random.randn(x.size)
y_noise = y + iid
```

```python
plt.figure(figsize=(15,6))
plt.plot(x,y_noise,'.k',label = "Data")
plt.plot(x,y,'b',lw=2,label ="The true Gaussian process")
plt.legend()
plt.show()
```
