```python

# Usually we use `%matplotlib inline`. However we need `notebook` for the anim to render in the notebook.
#%matplotlib notebook
%matplotlib inline


#import matplotlib
#import matplotlib.animation as animation
from IPython.display import clear_output
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from gpasdlm.kernels import *
```

```python
#simulation 1d (for time series modelling and forecasting)
```

```python

n=1000
x =np.linspace(-8,8,n)
kernels = [RBF(),Matern52(),Matern32(),Matern12(),Linear(),Periodic(period=2)]
```

```python
Names = list(map(lambda x: x.__class__.__name__,kernels))
#Names
```

```python
y_list =[]

for ker in kernels:
    gp_1d = ker
    y = gp_1d.simulate(x)
    y_list.append(y)
    

```

```python
    
fig, ax = plt.subplots(2,3,sharex =True, figsize=(12,5))
fig.suptitle("Gaussian Process 1d simulation") # 
k=0
for i in range(2):
    for j in range(3):
        ax[i,j].plot(x,y_list[k],'b')
        ax[i,j].set_title(Names[k].upper())
        k+=1
```

```python
fig, ax = plt.subplots(2,3,sharex =True, figsize=(12,5))
fig.suptitle("Gaussian Process 1d simulation") # 
k=0
y_list_noise = []
for i in range(2):
    for j in range(3):
        iid= np.random.randn(x.size)
        y_list_noise.append(y_list[k] + .5*iid)
        ax[i,j].plot(x,y_list_noise[k],'ko')
        ax[i,j].plot(x,y_list[k],'b')
        ax[i,j].set_title(Names[k].upper())
        k+=1
```

```python
#from gpbydl import KernelSum
#k = KernelSum(Periodic(),RBF())
```

```python
import gpbydl
```

```python
from gpbydl import GaussianProcessRegressor as GPR
k=0
ypred_list = []
for y in y_list_noise:
    model = GPR()
    model.kernel_choice = Names[k]
    model.fit(x,y)
    k+=1
    ypred,_=model.predict(x,y)
    ypred_list.append(ypred)

    
    
```

```python
fig, ax = plt.subplots(2,3,sharex =True, figsize=(12,5))
fig.suptitle("Gaussian Process 1d simulation") # 
k=0
for i in range(2):
    for j in range(3):
        ax[i,j].plot(x,y_list_noise[k],'ko')
        ax[i,j].plot(x,y_list[k],'b')
        ax[i,j].plot(x,ypred_list[k],'r')

        ax[i,j].set_title(Names[k].upper())
        k+=1
```

```python
Y_list =[]
x = np.linspace(-3,3,300)
kernels = [RBF(),Matern32(),Matern12(),Matern52(),Periodic(period=3),Linear()]
Names = list(map(lambda x: x.__class__.__name__,kernels))
Names
for ker in kernels:
    gp_2d = ker
    y = gp_2d.simulate_2d(x)
    Y_list.append(y)
    

    
```

```python
fig, ax = plt.subplots(2,3,sharex =True, figsize=(10,5))
fig.suptitle("Gaussian Process 2d simulation") # 
k=0
for i in range(2):
    for j in range(3):
        ax[i,j].imshow(Y_list[k], cmap ="jet")
        ax[i,j].set_title(Names[k].upper())
        k+=1

```
